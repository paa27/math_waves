```python
#linear parametric
N=10000
dt=0.01

wavep = Standing(mode=1,max_order=1,harmonics=1,
	params={'h':30000.0,'g0':1.0,'sigma':1.0,'k0':pi})

solp,ts,hamp = wavep.solve(dt=dt,N=N,x0=np.float64([0.00001,0]),
                           forcing={'epsilon':0.01,'delta':0.25,'j':1})
solp = np.float64([[solp[i][n] for i in range(len(solp))]for n in range(2*wavep.harmonics)])
```

```python
#linear parametric
lower=3000
upper=9000
env = np.abs(hilbert(solp[0]))
x = ts[lower:upper]
y = np.log(env)[lower:upper]
X = x - x.mean()
Y = y - y.mean()

slope = (X.dot(Y))/(X.dot(X))

sigma1 = slope/(wavep.force_eps*wavep.force_omega)

f,(ax1,ax2) = plt.subplots(1, 2,figsize=(20, 8))

ax1.plot(ts,solp[0],label=r'$a_{j}$')
ax1.plot(ts[lower:upper],env[lower:upper],label='envelope')
ax1.set_xlabel('t',size=20)
ax1.set_ylabel('a',size=20)
ax1.set_title('Fourier coeff (free surface, j=1)',size=20)
ax1.legend(prop={'size': 15})
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10)


ax2.plot(ts,hamp)
ax2.set_xlabel('t',size=20)
ax2.set_ylabel('h',size=20)
ax2.set_title('Hamiltonian',size=20)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10)
ax2.yaxis.offsetText.set_fontsize(10)
```

```python
#5th order
N=10000
dt=0.001 #RK4 error is O(dt^5)

tps = np.linspace(0.0,(N-1)*dt,num=N,dtype='float64')

wave5 = Standing(mode=1,max_order=5,harmonics=5,
	params={'h':30000.0,'g0':1.0,'sigma':0.0,'k0':pi})

wave = Standing(mode=1,max_order=1,harmonics=3,
	params={'h':30000.0,'g0':1.0,'sigma':0.0,'k0':pi})

A=0.01*wave._k(1) #Exact equation has error O(A^6)
sigma = sqrt(1 - 1/4.0*A**2-13/128.0*A**4)

ts = tps*(wave.g(1,0)*wave._k(1))**(0.5)

a1 = (A + 3/32.0*A**3-137/3072.0*A**5)*np.sin(sigma*ts)+(1/16.0*A**3-11/5376.0*A**5)*np.sin(3*sigma*ts)+\
163/21504.0*A**5*np.sin(5*sigma*ts)
a2 = 1/4.0*A**2+1/16.0*A**4-(1/4.0*A**2-25/192.0*A**4)*np.cos(2*sigma*ts)-67/1344.0*A**4*np.cos(4*sigma*ts)
a3 = (9/32.0*A**3-1/256.0*A**5)*np.sin(sigma*ts)-(3/32.0*A**3-2195/14336.0*A**5)*np.sin(3*sigma*ts)-\
16365/473088*A**5*np.sin(5*sigma*ts)
a4 = 1/8.0*A**4-1/6.0*A**4*np.cos(2*sigma*ts)+1/24.0*A**4*np.cos(4*sigma*ts)
a5 = 145/768.0*A**5*np.sin(sigma*ts)-515/3072.0*A**5*np.sin(3*sigma*ts)+85/3072.0*A**5*np.sin(5*sigma*ts)
a1 = 1/wave._k(1)*a1
a2 = 1/wave._k(1)*a2
a3 = 1/wave._k(1)*a3
a4 = 1/wave._k(1)*a4
a5 = 1/wave._k(1)*a5


a1_dot = sigma*((A + 3/32.0*A**3-137/3072.0*A**5)*np.cos(sigma*ts)+3*(1/16.0*A**3-11/5376.0*A**5)*np.cos(3*sigma*ts)+\
               5*163/21504.0*A**5*np.cos(5*sigma*ts))
a2_dot = sigma*(2*(1/4.0*A**2-25/192.0*A**4)*np.sin(2*sigma*ts)+4*(67/1344.0*A**4)*np.sin(4*sigma*ts))
a3_dot = sigma*((9/32.0*A**3-1/256.0*A**5)*np.cos(sigma*ts)-3*(3/32.0*A**3-2195/14336.0*A**5)*np.cos(3*sigma*ts)-\
               5*16365/473088*A**5*np.cos(5*sigma*ts))
a4_dot = sigma*(2*1/6.0*A**4*np.sin(2*sigma*ts)-4*1/24.0*A**4*np.sin(4*sigma*ts))
a5_dot = sigma*(145/768.0*A**5*np.cos(sigma*ts)-3*515/3072.0*A**5*np.cos(3*sigma*ts)+\
               5*85/3072.0*A**5*np.cos(5*sigma*ts))

a1_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a1_dot
a2_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a2_dot
a3_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a3_dot
a4_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a4_dot
a5_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a5_dot

b1 = a1_dot/(wave.k(1)*wave.T(1))
b2 = a2_dot/(wave.k(2)*wave.T(2))
b3 = a3_dot/(wave.k(3)*wave.T(3))
b4 = a4_dot/(wave.k(4)*wave.T(4))
b5 = a5_dot/(wave.k(5)*wave.T(5))
```

```python
sol5,_,h5 = wave5.solve(dt=dt,
                        N=N,
                        x0=np.float64([a1[0]/2.0,b1[0]/2.0,
                                       a2[0]/2.0,b2[0]/2.0,
                                       a3[0]/2.0,b3[0]/2.0,
                                       a4[0]/2.0,b4[0]/2.0,
                                       a5[0]/2.0,b5[0]/2.0]))
```

```python
sol5 = np.float64([[sol5[i][n] for i in range(len(sol5))]for n in range(2*wave5.harmonics)])
sol_exact = np.float64([a1,b1,a2,b2,a3,b3,a4,b4,a5,b5])/2.0
diff5 = np.abs(sol_exact-sol5)
np.save('./sol5.npy',sol5)
```

```python
#3rd order
N=40000
dt=0.001 #RK4 error is O(dt^5)

tps = np.linspace(0.0,(N-1)*dt,num=N,dtype='float64')

wave3 = Standing(mode=1,max_order=3,harmonics=3,
	params={'h':30000.0,'g0':1.0,'sigma':0.0,'k0':pi})

wave = Standing(mode=1,max_order=1,harmonics=3,
	params={'h':30000.0,'g0':1.0,'sigma':0.0,'k0':pi})

A=0.001*wave._k(1) #Exact equation has error O(A^4)
sigma = sqrt(1 - 1/4.0*A**2)

ts = tps*(wave.g(1,0)*wave._k(1))**(0.5)

a1 = (A + 3/32.0*A**3)*np.sin(sigma*ts)+(1/16.0*A**3)*np.sin(3*sigma*ts)
a2 = 1/4.0*A**2-(1/4.0*A**2)*np.cos(2*sigma*ts)
a3 = (9/32.0*A**3)*np.sin(sigma*ts)-(3/32.0*A**3)*np.sin(3*sigma*ts)

a1 = 1/wave._k(1)*a1
a2 = 1/wave._k(1)*a2
a3 = 1/wave._k(1)*a3


a1_dot = sigma*((A + 3/32.0*A**3)*np.cos(sigma*ts)+3*(1/16.0*A**3)*np.cos(3*sigma*ts))
a2_dot = sigma*(2*(1/4.0*A**2)*np.sin(2*sigma*ts))
a3_dot = sigma*((9/32.0*A**3)*np.cos(sigma*ts)-3*(3/32.0*A**3)*np.cos(3*sigma*ts))


a1_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a1_dot
a2_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a2_dot
a3_dot = wave.g(1,0)**(0.5)/wave._k(1)**(0.5)*a3_dot


b1 = a1_dot/(wave.k(1)*wave.T(1))
b2 = a2_dot/(wave.k(2)*wave.T(2))
b3 = a3_dot/(wave.k(3)*wave.T(3))
```

```python
sol3,_,h3 = wave3.solve(dt=dt,N=N,x0=np.float64([a1[0]/2.0,b1[0]/2.0,a2[0]/2.0,b2[0]/2.0,a3[0]/2.0,b3[0]/2.0]))
sol3 = np.float64([[sol3[i][n] for i in range(len(sol3))]for n in range(2*wave3.harmonics)])
sol_exact = np.float64([a1,b1,a2,b2,a3,b3])/2.0
diff3 = np.abs(sol_exact-sol3)
np.save('./sol3.npy',sol3)
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(20, 7))
f.subplots_adjust(wspace=0.5)

ax1.plot(tps,sol3[0],label=r'$a_{j}$')
ax1.plot(tps,sol3[1],label=r'$b_{j}$')
ax1.set_xlabel('t',size=20)
ax1.set_ylabel('a',size=20)
ax1.set_title('j',size=20)
ax1.legend(prop={'size': 15})
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10)

ax2.plot(tps,sol3[2],label=r'$a_{2j}$')
ax2.plot(tps,sol3[3],label=r'$b_{2j}$')
ax2.set_xlabel('t',size=20)
ax2.set_ylabel('a',size=20)
ax2.set_title('2j',size=20)
ax2.legend(prop={'size': 15})
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10)

ax3.plot(tps,sol3[4],label=r'$a_{3j}$')
ax3.plot(tps,sol3[5],label=r'$b_{3j}$')
ax3.set_xlabel('t',size=20)
ax3.set_ylabel('a',size=20)
ax3.set_title('3j',size=20)
ax3.legend(prop={'size': 15})
ax3.tick_params(axis='both', which='major', labelsize=12)
ax3.tick_params(axis='both', which='minor', labelsize=10)

f.suptitle('Numerical',size=20)
```
